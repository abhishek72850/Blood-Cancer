## Misc Functions

```python
def imageSegmentation(img):
    lower=np.array([80,100,190])
    upper=np.array([180,255,255])

    #hsv=cv2.cvtColor(nemo, cv2.COLOR_RGB2HSV)
    
    mask=cv2.inRange(img,lower,upper)
    res=cv2.bitwise_and(img,img,mask=mask)
    #ker=np.ones((15,15),np.float32)/255
    #smooth=cv2.filter2D(res,-1,ker)
    #blur=cv2.GaussianBlur(res,(15,15),0)
    median=cv2.medianBlur(res,15)
    
    return median,mask

def fractal_dimension(Z, threshold=0.9):

    # Only for 2d image
    assert(len(Z.shape) == 2)

    # From https://github.com/rougier/numpy-100 (#87)
    def boxcount(Z, k):
        S = np.add.reduceat(
            np.add.reduceat(Z, np.arange(0, Z.shape[0], k), axis=0),
                               np.arange(0, Z.shape[1], k), axis=1)

        # We count non-empty (0) and non-full boxes (k*k)
        return len(np.where((S > 0) & (S < k*k))[0])


    # Transform Z into a binary array
    Z = (Z < threshold)

    # Minimal dimension of image
    p = min(Z.shape)

    # Greatest power of 2 less than or equal to p
    n = 2**np.floor(np.log(p)/np.log(2))

    # Extract the exponent
    n = int(np.log(n)/np.log(2))

    # Build successive box sizes (from 2**n down to 2**1)
    sizes = 2**np.arange(n, 1, -1)

    # Actual box counting with decreasing size
    counts = []
    for size in sizes:
        counts.append(boxcount(Z, size))

    # Fit the successive log(sizes) with log (counts)
    coeffs = np.polyfit(np.log(sizes), np.log(counts), 1)
    return -coeffs[0]

def getContours(median):
    img=cv2.cvtColor(median, cv2.COLOR_RGB2GRAY)

    ret,thresh = cv2.threshold(img,100,255,0)

    contours, hierarchy = cv2.findContours(thresh, 1,2)
    
    img2 = cv2.drawContours(img, contours, -1, (255,255,255), 3)
    
    return thresh,contours,img,img2

def extractPhysicalFeature(median,mask):
    
    thresh,contours,img,img2 = getContours(median)
    
    contours = sorted(contours,key=len)
    #M = cv2.moments(cnt)

    #epsilon = 0.1*cv2.arcLength(cnt,True)
    #approx = cv2.approxPolyDP(cnt,epsilon,True)
    
    #Area
    if(len(contours) > 0):
        area = cv2.contourArea(contours[-1])
        if(area != 0):
            area_err = 0.5/area
        else:
            area_err = 0
    else:
        area = 0
        area_err = 0
    
    #Perimeter
    if(len(contours) > 0):
        perimeter = cv2.arcLength(contours[-1],True)
        if(perimeter != 0):
            perimeter_err = 0.5/perimeter
        else:
            perimeter_err = 0
    else:
        perimeter = 0
        perimeter_err = 0
    
    #Radius
    equi_diameter = np.sqrt(4*area/np.pi)
    if(equi_diameter != 0):
        radius = equi_diameter/2
        radius_err = 0.5/radius
    else:
        radius = 0
        radius_err = 0

    #Smoothness
    smoothness = np.average(np.absolute(ndi.filters.laplace(median.astype(float) / 255.0)))
    if(smoothness != 0):
        smoothness_err = 0.5/smoothness
    else:
        smoothness_err = 0
    
    #Compactness
    compactness = np.square(perimeter)/(area-1.0)
    if(compactness != 0):
        compactness_err = 0.5/compactness
    else:
        compactness_err = 0

    #texture
    texture = max(cv2.mean(img,mask = mask))
    #print(texture)
    if(texture != 0):
        texture_err = 0.5/texture
    else:
        texture_err = 0

    #Concave Points
    if(len(contours) > 0):
        hull = cv2.convexHull(contours[-1],returnPoints = False)
        defects = cv2.convexityDefects(contours[-1],hull)
        #print(type(defects))
        if(defects is not None):
            #print(defects)
            concave_points=len(defects)
            concave_points_err = 0.5/concave_points
        else:
            concave_points = 0
            concave_points_err = 0
    else:
        hull = 0
        defects = 0
        #print(contours,defects)
        concave_points = 0
        concave_points_err = 0
    
    #Fractal Dimension
    f_dim = fractal_dimension(img)
    f_dim_err = 0.5/f_dim
    
    return [radius,texture,perimeter,area,smoothness,compactness,concave_points,f_dim,
            radius_err,texture_err,perimeter_err,area_err,smoothness_err,compactness_err,concave_points_err,f_dim_err]

def saveModel(model):
    model_json = model.to_json()
    with open("model.json", "w") as json_file:
        json_file.write(model_json)

    # serialize weights to HDF5
    model.save_weights("model.h5")
    print("Saved model to disk")

def loadModel():
    # load json and create model
    json_file = open('model.json', 'r')
    loaded_model_json = json_file.read()
    json_file.close()
    loaded_model = model_from_json(loaded_model_json)
    # load weights into new model
    loaded_model.load_weights("model.h5")
    print("Loaded model from disk")
    
    loaded_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return loaded_model
```

## Preparing Dataset from images

```python
f_lst=[]

for root, dirs, files in os.walk("dataset2-master/dataset2-master/images"):
    #char_list.append(os.path.basename(root).split('_')[-1])
    for f in files:
        features=[]
        img = cv2.imread(os.path.join(root,f))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        median,mask = imageSegmentation(img)
        print(root,f)
        cv2.imwrite("new_segmented_dataset/"+f.split('.')[0].split('_')[-1]+".png",median)
        
        features.append(int(f.split('.')[0].split('_')[-1]))
        features.extend(extractPhysicalFeature(median,mask))
        features.append(root.split('\\')[-1])
        
        f_lst.append(features)
        
f_df = pd.DataFrame(f_lst,columns=['id','radius','texture','perimeter','area',
                                   'smoothness','compactness','concave_points','f_dim',
                                    'radius_err','texture_err','perimeter_err','area_err',
                                   'smoothness_err','compactness_err','concave_points_err',
                                   'f_dim_err','category'])
```

```python
f_df.reset_index()
f_df.fillna(0,inplace = True)

lb_make = LabelEncoder()
f_df["category_code"] = lb_make.fit_transform(f_df["category"])

f_df.to_csv('new_dataset.csv',index=None)
f_df.head()
```
